### Constructing a DataFrame using dict

csv 파일을 불러와서 data frame 변수로 바꾸었는데, 처음부터 정보를 가지고 여러가지를 만들어 쓸 수는 없을까? dict는 중괄호로 만들게 된다. 

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df 
```

dict안에 5가지의 내용이 들어가게 된다. dict는 key와 value로 구분이 된다. value에 list의 형태가 들어가 있다. 5가지의 아이템이 들어가게 된다. pd.dataframe에 dict라는 데이터를 넣어주면, dataframe을 만들어주는 것이다. csv파일을 통해 불러올 수도 있지만, dict를 통해, 일일히 정보를 적어주고 불러올 수도 있다. df를 dataframe 변수라고 한다. 액셀 파일과 똑같은 것을 알 수 있다. 

### Constructing a DataFrame using pd.Series

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

dataframe을 만드는 또다른 방법도 있다. pd.Series를 활용하면 된다. 아까는 dict를 먼저 만들었다면, 이번에는 series로 내용만 5가지 만들고, 다시 함수를 쓸 때, dict를 써주면 된다. 
다시 요약하면, csv파일을 불러와서 변환하는 과정이 있고, dict파일을 통으로 만들어서 dataframe을 만들 수도 있고, 마지막은 series를 활용해서 column의 내용만 따로 만들고, dict를 다시 활용하는 방법이 있다. 가장 흔한 방법은 csv파일이다. 

### Changing column names

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

list의 경우처럼, 특정 정보만 접근할 수도 있다. 일부의 내용을 변경, 수정하는 경우도 있는데, 이러한 것을 배울 것이다. df.columns을 입력하게 되면, column에 해당하는 그 정보만 나오게 된다. species 부분을 소문자로 바꾸는 경우가 나와있다. 이런 식으로 column 명을 바꿀 수 있다. 

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']
```

loc는 location이다. 콜론이 들어가면, 전부다라는 의미이다. 행에 대해서 전부다라는 것이다. 

```python
# Access one column (short-hand)
df['sepal_length']
```

column 명만 적어 주어도 같은 결과값이 나올 수도 있다. 

```python
# Access one column (shortest-hand)
df.sepal_length
```

더 간단하게 나타낼 수도 있다. 

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

두개의 column을 가져올 수도 있다. 

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

더 간단하게 나타낸 것이다. 

### Accessing rows

```python
# Access one row 
df.loc[3, :]
```

row 값을 가져올 때도 마찬가지이다. 3 번째 row와 콜론을 적어주는 것이다. 

```python
# Access two rows
df.loc[[3, 0], :]
```

두개의 row 정보를 가져올 수 도 있다. list로 써주면 된다. 

```python
# Access a range of rows
df.loc[2:5, :]
```

2에서 4번째까지 row정보를 가져오는 것이다. 

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]
```

sepal length가 5보다 큰 경우의 정보만을 가져오는 경우이다. 

### Accessing a cell

하나의 셀만을 가져오고 싶은 경우도 있다.

4번째의 셀의 정보만을 가져오는 경우이다.

더 간단하게 나타낸 경우이다.

### Adding/Removing new columns

```python
df['new_column'] = 1
df
```

새로운 column이 생겼다. 

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

새로운 column에 1,2,3,4,5 5가지가 들어가있다.

```python
df = df.drop('new_column', axis=1) # axis=1: column
df
```

0은 행정보이고, 1은 열의 정보이다. 

### Adding/Removing new rows

```python
df.loc[5, :] = ''
df
```

string인데, 그 안에 아무 것도 없다. 아무 것도 없는 string이다. 5번째 row로 만든 것이다. 

```python
df = df.drop(5, axis=0) # axis=0: row
df
```

5를 drop하는데 행축인지, 열축인지 모르기 때문에 행축(0)에서 찾으라는 것이다.
